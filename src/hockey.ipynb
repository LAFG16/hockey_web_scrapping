```python
import pandas as pd 
import numpy as np
import time
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec 
from selenium.common.exceptions import StaleElementReferenceException
```

```python
driver = webdriver.Chrome()
url = "https://www.scrapethissite.com/pages/forms/?page_num=1"
driver.get(url)
```

```python
driver.find_element(By.ID, "per_page").click()
driver.find_element(By.XPATH, "//option[@value='100']").click()
```

```python
driver.find_element(By.ID, "per_page").click()
driver.find_element(By.XPATH, "//option[@value='100']").click()
driver.find_element(By.XPATH, "//a[.//span[text()='»']]").click()
```

```python
name = []
year = []
wins = []
losses = []
ot_losses = []
win_percent = []
goals_for = []
goals_against = []
more_less = []

while True:
    try:
        # Espera a que la página se cargue completamente antes de continuar
        WebDriverWait(driver, 10).until(
            ec.presence_of_element_located((By.CLASS_NAME, "team"))
            )

        table = driver.find_elements(By.CLASS_NAME, "team")
        for i in table:
            name_i = [name.append(n.text) for n in i.find_elements(By.CLASS_NAME, "name")]
            year_i = [year.append(n.text) for n in i.find_elements(By.CLASS_NAME, "year")]
            wins_i = [wins.append(n.text) for n in i.find_elements(By.CLASS_NAME, "wins")]
            losses_i = [losses.append(n.text) for n in i.find_elements(By.CLASS_NAME, "losses")]
            ot_losses_i = [ot_losses.append(n.text) for n in i.find_elements(By.CLASS_NAME, "ot-losses")]
            win_porcent_i = [win_percent.append(n.text) for n in i.find_elements(By.CSS_SELECTOR, ".pct.text-success, .pct.text-danger")]
            goals_for_i = [goals_for.append(n.text) for n in i.find_elements(By.CLASS_NAME, "gf")]
            goals_against_i = [goals_against.append(n.text) for n in i.find_elements(By.CLASS_NAME, "ga")]
            more_less_i = [more_less.append(n.text) for n in i.find_elements(By.CSS_SELECTOR, ".diff.text-success, .diff.text-danger")]

        next_button = driver.find_element(By.XPATH, '//a[@aria-label="Next"]')

        next_button.click()
    except StaleElementReferenceException:
        # Si se produce la excepción, intenta nuevamente
        pass
    except Exception as e:
        # Si hay algún otro error, finaliza el bucle
        print(f"Error: {str(e)}")
        break
    
```

```python
hockey_df = pd.DataFrame({
    "Name": name,
    "Year": year,
    "Wins": wins,
    "Losses": losses,
    "OT Losses": ot_losses,
    "Win Percentage": win_percent,
    "Goals For": goals_for,
    "Goals Agaeinst": goals_against,
    "+ / -": more_less    
})
hockey_df
```
